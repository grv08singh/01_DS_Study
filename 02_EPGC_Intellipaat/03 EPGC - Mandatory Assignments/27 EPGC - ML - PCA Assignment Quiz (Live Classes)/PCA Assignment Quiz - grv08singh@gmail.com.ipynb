```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
d = load_digits()
```

```python
d['data']
```

```python
df = pd.DataFrame(d['data'], columns=d['feature_names'])
```

```python
df.head(2)
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
df_copy = df.copy()
```

```python
sc = StandardScaler()
```

```python
df = pd.DataFrame(sc.fit_transform(df), columns=df_copy.columns)
```

```python
df.head(2)
```

```python
pca = PCA()
```

```python
pca.fit_transform(df)
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_.max()
```

```python
variance = pca.explained_variance_ratio_.cumsum()
```

```python
features = np.arange(64)
```

```python
plt.figure(figsize=(15,5))
plt.plot(features,variance)
plt.plot(features,np.ones(64)*0.9)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance')
plt.show()
```

```python
variance
```

#### from the plot, we can see, there are 30 Principal Components that explain 90% variance in the data
